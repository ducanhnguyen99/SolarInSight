## Preproprocess Data

Sort each dataframe based on the timestamp column.

```python
import pandas as pd
dataframes = [df2006, df2007, df2008, df2009, df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022, df2023]
updated_dataframes = []

for df in dataframes:
  updated_df = df[pd.notnull(df['Datum__MEZ_'])]
  updated_df = updated_df.sort_values(by="Datum__MEZ_")

  updated_dataframes.append(updated_df)

df2006_updated = updated_dataframes[0]
df2007_updated = updated_dataframes[1]
df2008_updated = updated_dataframes[2]
df2009_updated = updated_dataframes[3]
df2010_updated = updated_dataframes[4]
df2011_updated = updated_dataframes[5]
df2012_updated = updated_dataframes[6]
df2013_updated = updated_dataframes[7]
df2014_updated = updated_dataframes[8]
df2015_updated = updated_dataframes[9]
df2016_updated = updated_dataframes[10]
df2017_updated = updated_dataframes[11]
df2018_updated = updated_dataframes[12]
df2019_updated = updated_dataframes[13]
df2020_updated = updated_dataframes[14]
df2021_updated = updated_dataframes[15]
df2022_updated = updated_dataframes[16]
df2023_updated = updated_dataframes[17]



```

In df2018 connect the columns *Day_Ahead_Auktion__DE_AT_LU_*and *Day_Ahead_Auktion__DE_LU_* to get one *Day_Ahead_Auktion* column.

```python
df2018_updated["Day_Ahead_Auktion"] = None

df2018_updated.loc[df2018_updated["Day_Ahead_Auktion__DE_AT_LU_"].notnull(), "Day_Ahead_Auktion"] = df2018_updated["Day_Ahead_Auktion__DE_AT_LU_"]
df2018_updated.loc[df2018_updated["Day_Ahead_Auktion__DE_LU_"].notnull(), "Day_Ahead_Auktion"] = df2018_updated["Day_Ahead_Auktion__DE_LU_"]

df2018_updated = df2018_updated.drop(["Day_Ahead_Auktion__DE_AT_LU_","Day_Ahead_Auktion__DE_LU_"], axis=1)

df2018_updated
```

## Create ARIMA Model

### Train the model

Train the model and save it in BigQuery in electricity_data under models as day_ahead_auktion_forecast.

*As this cell was executed the model is already saved in BigQuery, therefore this cell doesn't need to be executed anymore for the following cells to work.*

```python
# Train ARIMA Model

%%bigquery df_arima --project solarinsight-383513
CREATE OR REPLACE MODEL electricity_data.day_ahead_auktion_forecast
OPTIONS(
  model_type = 'ARIMA_PLUS',
  time_series_timestamp_col = 'Datum__MEZ_',
  time_series_data_col = 'Day_Ahead_Auktion',
  auto_arima = TRUE,
  data_frequency = 'AUTO_FREQUENCY',
  decompose_time_series = TRUE,
  horizon = 10000,
  auto_arima_max_order = 5,
  holiday_region='DE'
)
AS
SELECT
  Datum__MEZ_,
  Day_Ahead_Auktion

FROM
 `solarinsight-383513.electricity_data.day_ahead_auktion`
```

### Evaluate the Model with Big Query

```python
%%bigquery df_evaluate --project solarinsight-383513
SELECT
 *
FROM
 ML.ARIMA_EVALUATE(MODEL electricity_data.day_ahead_auktion_forecast)

```

Look at model evaluation: BiqQuery ML trained 42 models and selected the model with the lowest AIC as the final model. In our case 703127.536145.

```python
df_evaluate
```

Look at the Model Coefficients with Big Query

```python
%%bigquery df_coef --project solarinsight-383513
SELECT
 *
FROM
 ML.ARIMA_COEFFICIENTS(MODEL electricity_data.day_ahead_auktion_forecast)
```

Print coefficients. *Looking at the coefficents we see that the ar_coefficients part is empty. Meaning that our model don't have an autoregressiv part.*

```python
df_coef
```

Use BigQuery to join the four data frames and save the concatenated data frame in *electricity_production_data* as *weather_forecast_data* in BigQuery.

```python
%%bigquery df_forecast_weather --project solarinsight-383513
CREATE OR REPLACE TABLE electricity_production_data.weather_forecast_data AS(
SELECT t1.STATIONS_ID,t1.MESS_DATUM,t1.Wind_Speed,t2.Visibility,t3.AmbientTemp,t4.Humidity
FROM `solarinsight-383513.electricity_production_data.wind_speed_forecast_data` AS t1
INNER JOIN `solarinsight-383513.electricity_production_data.visibility_forecast_data` AS t2 ON t1.STATIONS_ID = t2.STATIONS_ID AND t1.MESS_DATUM=t2.MESS_DATUM
INNER JOIN `solarinsight-383513.electricity_production_data.AmbientTemp_forecast_data` AS t3 ON t2.STATIONS_ID = t3.STATIONS_ID AND t2.MESS_DATUM=t3.MESS_DATUM
INNER JOIN `solarinsight-383513.electricity_production_data.Humidity_forecast_data` AS t4 ON t3.STATIONS_ID = t4.STATIONS_ID AND t3.MESS_DATUM = t4.MESS_DATUM)
```
